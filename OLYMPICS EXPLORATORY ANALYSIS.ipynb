```python
#TOTAL NUMBER OF MEDALS
athletes_df.Medal.value_counts()
```

```python
#total number of athletes in each olympics
female_participants = athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')][['Sex', 'Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
male_participants = athletes_df[(athletes_df.Sex=='M') & (athletes_df.Season=='Summer')][['Sex','Year']]
male_participants = male_participants.groupby('Year').count().reset_index()
male_participants.tail()
```

```python
women_olympics = athletes_df[(athletes_df.Sex =='F') & (athletes_df.Season == 'Summer')]
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20, 10))
sns.countplot(x='Year', data=women_olympics, palette = 'Spectral')
plt.title('women participation')
```

```python
part = women_olympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20, 10))
part.loc[:, 'F'].plot()
plt.title('Plot of female athletes over time')
```

```python
#GOLD MEDAL ATHLETES
goldmedals = athletes_df[(athletes_df.Medal == 'Gold')]
goldmedals.head()
```

```python
goldmedals = goldmedals[np.isfinite(goldmedals['Age'])] 
```

```python
#GOLD MEDALS BEYOND 60 YEARS OLD
goldmedals['ID'][goldmedals['Age'] > 60].count()
```

```python
sporting_event = goldmedals['Sport'][goldmedals['Age']>60]
sporting_event
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('GOLD MEDALS FOR ATHLETES ABOVE 60')
```

```python
#GOLD MEDALS FOR EACH COUNTRY 
goldmedals.Region.value_counts().reset_index(name = 'Medals').head(5)
```

```python
totalGoldMedals = goldmedals.Region.value_counts().reset_index(name='Medals').head(6)
g = sns.catplot(x='index', y='Medals', data=totalGoldMedals,
               height=5, kind='bar', palette='rocket')
g.despine(left=True)
g.set_xlabels('Top 5 countries')
g.set_ylabels('Number of Medals')
plt.title('gold medals per country')
```
